### I will make a new DataFrame to see the differences between the 30 cities in the study. 

```python
City_MH_ = pd.DataFrame(columns=['city', 'Stu_pop', 'student %', 'depressive %', 'dep students %','dep workers %'])
```

```python
City_MH_
```

```python
def city_life(live,n):
    dep = Mental_Health[Mental_Health['City'] == live].Depression.sum()
    dep_stu = Mental_Health[(Mental_Health.City == live)&(Mental_Health.Work == 'Student')].Depression.sum()
    num_stu = Mental_Health[(Mental_Health.City == live)&(Mental_Health.Work == 'Student')].Depression.count()
    dep_wor = Mental_Health[(Mental_Health.City == live)&(Mental_Health.Work == 'Working Professional')].Depression.sum()
    num_wor = Mental_Health[(Mental_Health.City == live)&(Mental_Health.Work == 'Working Professional')].Depression.count()
    city_pop = Mental_Health[Mental_Health['City'] == live].Depression.count()
    p_stu = round(((num_stu/city_pop)*100),2) # percentage of students in the study (I suppose that is the same in the city  
    p_dep = round(((dep/city_pop)*100),2) # percentage of the people with depression 
    p_d_stu = round(((dep_stu/num_stu)*100),2) # percentage of depressive students
    p_d_wor = round(((dep_wor/num_wor)*100),2) # percentage of depressive workers


    City_MH_.loc[n] = [live, city_pop, p_stu, p_dep, p_d_stu, p_d_wor] 

    return City_MH_
```

```python
# I made a Python list because I wanted to see the values of the different cities.  
lista_ciudades = city_val.tolist()
lista_ciudades
```

```python
n=0
for i in lista_ciudades:
    city_life(i,n)
    n = n+1
City_MH_
```

### All the erroneous values are in a new city, 'Lost_city'

```python
Mental_Health['City'] = Mental_Health['City'].replace(to_replace=change_miss, value='Lost_city')
```

### Figures of depression in the 31 cities of the study.

```python
for city in Mental_Health['City'].unique():
    city_df = Mental_Health[Mental_Health['City'] == city]

    # Create separate plots for males and females
    plt.figure(figsize=(12, 4))

    # Plot for males
    plt.subplot(1, 2, 1)
    sns.histplot(data=city_df[city_df['Gender'] == 'Male'], x='Age', binwidth=2, hue='Depression', kde=True)
    plt.title(f'Age Distribution of Males in {city}')

    # Plot for females
    plt.subplot(1, 2, 2)
    sns.histplot(data=city_df[city_df['Gender'] == 'Female'], x='Age',binwidth=2, hue='Depression', kde=True)
    plt.title(f'Age Distribution of Females in {city}')

    plt.tight_layout()
    plt.savefig(f'fig/City_{city}.png')
    plt.show()
```

# The data for the 30 cities shows different populations with depression. 

- The city has a distinct nonidentifiable effect on the population depression.
- In the city, the population affected by the depression changes smoothly with age.
- The city has a more significant effect on the depression than the student-worker separation 
- There are small differences in depression in the gender of the people in every city.
