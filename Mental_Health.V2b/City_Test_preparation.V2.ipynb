```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Mental_Health_Test = pd.read_csv('Datasheet/Mental_Health_test.csv')
```

```python
Mental_Health_Test.head(3)
```

```python
def horas_cat(value):
    if value in ["Less than 5 hours", "3-4 hours", "1-2 hours", "2-3 hours","4-5 hours"]:
        return 4.5
    elif value in ["5-6 hours"]:
        return 5.5
    elif value in ["6-7 hours", "6-8 hours"]:
        return 6.5
    elif value in ["7-8 hours"]:
        return 7.5
    elif value in [ "More than 8 hours"]:
        return 8.5
    elif value in [ "8-9 hours", "9-10 hours", "9-11 hours", "10-11 hours"]:
        return 10
    elif value in [ "50-75 hours", "55-66 hours", "60-65 hours", "20-21 hours", "35-36 hours","40-45 hours",
                   "45-48 hours"]:
        return 12 
    else:
        return np.nan
```

```python
Mental_Health_Test['Sleep Duration'] = Mental_Health_Test['Sleep Duration'].apply(horas_cat)
```

```python
Mental_Health_Test['Sleep Duration']= Mental_Health_Test['Sleep Duration'].fillna(Mental_Health_Test['Sleep Duration'].mean())
```

```python
Mental_Health_Test['Sleep Duration'].value_counts()
```

```python
    def Comida(value):
        if value in ['Healthy', "More Healthy"]:
            return 2
        elif value in ['Unhealthy']:
            return 0
        elif value in ['Moderate']:
            return 1
        else:
            return np.nan# Modules
```

```python
Mental_Health_Test['Dietary Habits'] = Mental_Health_Test['Dietary Habits'].apply(Comida)
```

```python
Mental_Health_Test['Dietary Habits']=Mental_Health_Test['Dietary Habits'].fillna(Mental_Health_Test['Dietary Habits'].mean())
```

```python
Mental_Health_Test['Dietary Habits'].value_counts()
```

```python
Mental_Health_Test = Mental_Health_Test.drop(columns = ['Name', 'Profession', 'CGPA', 'Degree'])
```

```python
Mental_Health_Test.head(3)
```

```python
Mental_Health_Test['Work'] = Mental_Health_Test['Working Professional or Student'] #change the name of the column
Mental_Health_Test.drop(columns='Working Professional or Student',axis=1, inplace=True) 
Mental_Health_Test.head(3)
```

```python
#drop 'Nan' to make the new columns
Mental_Health_Test['Academic Pressure'] = Mental_Health_Test['Academic Pressure'].replace(to_replace=np.nan, value=0)
Mental_Health_Test['Work Pressure'] = Mental_Health_Test['Work Pressure'].replace(to_replace=np.nan, value=0)
Mental_Health_Test['Study Satisfaction'] = Mental_Health_Test['Study Satisfaction'].replace(to_replace=np.nan, value=0)
Mental_Health_Test['Job Satisfaction'] = Mental_Health_Test['Job Satisfaction'].replace(to_replace=np.nan, value=0)
# Female = 1, Male = 0 
Mental_Health_Test['Gender'] = Mental_Health_Test['Gender'].replace(to_replace={'Female': 1, 'Male': 0})
#Yes = 1, No = 0
Mental_Health_Test['Have you ever had suicidal thoughts ?'] = Mental_Health_Test['Have you ever had suicidal thoughts ?'].replace(to_replace={'Yes': 1, 'No': 0})
Mental_Health_Test['Family History of Mental Illness'] = Mental_Health_Test['Family History of Mental Illness'].replace(to_replace={'Yes': 1, 'No': 0})
# Worker=1 Student = 0
# Mental_Health_Test['Work'] = Mental_Health_Test['Work'].replace(to_replace={'Working Professional': 1, 'Student': 0})
Mental_Health_Test.tail(3)
```

```python
Mental_Health_Test['Satisfaction'] = Mental_Health_Test['Study Satisfaction']+ Mental_Health_Test['Job Satisfaction']
```

```python
Mental_Health_Test['Satisfaction'].value_counts()
```

```python
Mental_Health_Test['Satisfaction'].replace(to_replace=6 , value=5,inplace=True)
```

```python
Mental_Health_Test = Mental_Health_Test.drop(columns=['Study Satisfaction', 'Job Satisfaction'],axis=1)
Mental_Health_Test['Pressure'] = Mental_Health_Test['Academic Pressure']+ Mental_Health_Test['Work Pressure']
Mental_Health_Test.head(3)
```
