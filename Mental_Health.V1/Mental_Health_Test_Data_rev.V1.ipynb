# Praparate the TEST set
- In this set, the transformations are the same as in the TRAIN dataset.
- The TEST dataset doesn't have the column depression. 

```python
Mental_Health = pd.read_csv('Datasheet/Mental_Health_test.csv')
```

```python

Mental_Health
```

```python
Mental_Health.info()
```

```python
Mental_Health = Mental_Health.drop(columns=['Name', 'City','Working Professional or Student','Profession', 'CGPA', 'Sleep Duration', 'Dietary Habits','Degree'],axis=1)
```

```python
Mental_Health
```

```python
Mental_Health['Academic Pressure'].replace(to_replace=np.nan, value=0,inplace=True)
```

```python
Mental_Health['Work Pressure'].replace(to_replace=np.nan, value=0,inplace=True)
```

```python
Mental_Health['Pressure'] = Mental_Health['Academic Pressure']+ Mental_Health['Work Pressure']
```

```python
Mental_Health = Mental_Health.drop(columns=['Work Pressure', 'Academic Pressure'],axis=1)
```

```python
Mental_Health
```

```python
Mental_Health['Study Satisfaction'].replace(to_replace=np.nan, value=0,inplace=True)
```

```python
Mental_Health['Job Satisfaction'].replace(to_replace=np.nan, value=0,inplace=True)
```

```python
Mental_Health['Satisfaction'] = Mental_Health['Study Satisfaction']+ Mental_Health['Job Satisfaction']
```

```python
Mental_Health['Satisfaction'].value_counts()
```

```python
Mental_Health['Satisfaction'].replace(to_replace=6 , value=5,inplace=True)
```

```python
Mental_Health
```

```python
Mental_Health = Mental_Health.drop(columns=['Study Satisfaction', 'Job Satisfaction'],axis=1)
```

```python
Mental_Health['Gender'].replace(to_replace={'Female': 1, 'Male': 0},inplace=True)
```

```python
Mental_Health['Have you ever had suicidal thoughts ?'].replace(to_replace={'Yes': 1, 'No': 0},inplace=True)
```

```python
Mental_Health['Family History of Mental Illness'].replace(to_replace={'Yes': 1, 'No': 0},inplace=True)
```
