# Eliminating columns 

```python
# eliminate columns
# 'Name', 'City', 'Working Professional or Student', 'Profession', 'CGPA', 'Sleep Duration', 'Dietary Habits','Degree',

# sum columns
# 'Work Pressure' + 'Academic Pressure'
#'Study Satisfaction' + 'Job Satisfaction'


Mental_Health = Mental_Health.drop(columns=['Name', 'City','Working Professional or Student','Profession', 'CGPA', 'Sleep Duration', 'Dietary Habits','Degree'],axis=1)
```

# Sum columns

```python
# sum columns
'Work Pressure' + 'Academic Pressure'
'Study Satisfaction' + 'Job Satisfaction'
```

```python
Mental_Health['Work Pressure'].value_counts()
```

```python
Mental_Health['Academic Pressure'].value_counts()
```

```python
Mental_Health['Academic Pressure'].replace(to_replace=np.nan, value=0,inplace=True)
```

```python
Mental_Health['Work Pressure'].replace(to_replace=np.nan, value=0,inplace=True)
```

```python
Mental_Health['Pressure'] = Mental_Health.apply(lambda x: x['Academic Pressure'] + x['Work Pressure'], axis=1)
```
